```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
data = {
    "Title": [
        "A Light in the Attic",
        "Tipping the Velvet",
        "Soumission",
        "Sharp Objects",
        "Sapiens",
        "The Hobbit",
        "Harry Potter",
        "The Alchemist",
        "Atomic Habits",
        "Rich Dad Poor Dad"
    ],
    "Price": [51.77, 53.74, 50.10, 47.82, 54.23, 35.99, 42.50, 29.99, 39.99, 24.99],
    "Availability": ["In stock"] * 10,
    "Rating": ["Three", "One", "One", "Four", "Five", "Four", "Five", "Three", "Five", "Four"]
}

df = pd.DataFrame(data)

print(df.head())

print("="*50)
print("FIRST 5 RECORDS")
print("="*50)
print(df.head())

# Dataset Information
print("\n" + "="*50)
print("DATASET SHAPE")
print("="*50)
print(df.shape)

print("\n" + "="*50)
print("COLUMN NAMES")
print("="*50)
print(df.columns)

print("\n" + "="*50)
print("DATA TYPES")
print("="*50)
print(df.dtypes)

print("\n" + "="*50)
print("DATASET INFO")
print("="*50)
print(df.info())

# Convert Rating to Numeric
rating_map = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5
}

df['Rating'] = df['Rating'].map(rating_map)

# Missing Values
print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)
print(df.isnull().sum())

# Statistical Summary
print("\n" + "="*50)
print("STATISTICAL SUMMARY")
print("="*50)
print(df.describe())

# Average Price
avg_price = df['Price'].mean()

print("\nAverage Price:", round(avg_price, 2))

# Most Common Rating
most_common_rating = df['Rating'].mode()[0]

print("Most Common Rating:", most_common_rating)

# Correlation
print("\nCorrelation Matrix")
print(df[['Price', 'Rating']].corr())

# -----------------------------
# Visualizations
# -----------------------------

sns.set_style("whitegrid")

# 1. Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=5, kde=True, color='skyblue')
plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# 2. Rating Distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Rating', data=df, hue='Rating', legend=False)
plt.title("Rating Distribution")
plt.show()

# 3. Boxplot for Outliers
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Price'])
plt.title("Price Outlier Detection")
plt.show()

# 4. Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Rating', y='Price', data=df)
plt.title("Price vs Rating")
plt.show()

# 5. Heatmap of Correlation
plt.figure(figsize=(5,4))
sns.heatmap(
    df[['Price','Rating']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()

print("\nEDA COMPLETED SUCCESSFULLY!")
```
